# sklearn-porter

Repository: https://github.com/nok/sklearn-porter

## AdaBoostClassifier

Documentation: [sklearn.ensemble.AdaBoostClassifier](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostClassifier.html)

### Loading data:

```python
from sklearn.datasets import load_iris

iris_data = load_iris()
X = iris_data.data
y = iris_data.target

print(X.shape, y.shape)
```

### Train classifier:

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

base_estimator = DecisionTreeClassifier(max_depth=4, random_state=0)
clf = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=100,
                         random_state=0)
clf.fit(X, y)
```

### Transpile classifier:

```python
%%time

from sklearn_porter import Porter

porter = Porter(clf, language='c')
output = porter.export()

with open('AdaBoostClassifier.c', 'w') as f:
    f.write(output)
```

```python
%%bash

cat AdaBoostClassifier.c
```

### Run classification in C:

```python
%%bash

gcc ./AdaBoostClassifier.c -std=c99 -lm -o AdaBoostClassifier
./AdaBoostClassifier 1 2 3 4
```
